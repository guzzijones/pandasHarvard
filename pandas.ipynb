```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/cs109/2014_data/master/diamonds.csv'
df = pd.read_csv(url,sep=',',index_col=0)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.carat.hist()
```

```python
#df.price.hist(bins=50)
plt.figure(1)
plt.hist(df.price,bins=50)
plt.title('distribution of price')
plt.xlabel('price')
```

```python
plt.figure(2,figsize=(10,5)) #size in incches
plt.hist(df['carat'],bins=200)
plt.title('dist of carat size')
plt.title('carat')
plt.xlim(0,3)
plt.ylim(0,5000)
```

```python
df.plot(x='carat',y='price',kind='scatter',color='black',alpha=.05,figsize=(10,5))
#fix the overplotting with the alpha value.  lower means less transparency
```

```python
plt.figure(2,figsize=(10,6))
plt.scatter(df['carat'],df['price'],color='black',alpha=.05)
plt.xlabel("carat")
plt.ylabel("price")
```

```python
df.groupby('color').plot(kind='scatter',x='carat',y='price')
```

```python
df.boxplot('price',by='color')
```

```python
df.boxplot('price',by='cut')
```
